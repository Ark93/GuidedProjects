```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(3)
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
yes_no = {
    "Yes" : True,
    "No" : False
}
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
star_wars["Which of the following Star Wars films have you seen? Please select all that apply."].fillna("No",inplace = True)
star_wars["Unnamed: 4"].fillna("No",inplace = True)
star_wars["Unnamed: 5"].fillna("No",inplace = True)
star_wars["Unnamed: 6"].fillna("No",inplace = True)
star_wars["Unnamed: 7"].fillna("No",inplace = True)
star_wars["Unnamed: 8"].fillna("No",inplace = True)
```

```python
movies = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones" : True,
    'Star Wars: Episode III  Revenge of the Sith' : True,
    "Star Wars: Episode IV  A New Hope" : True,
    "Star Wars: Episode V The Empire Strikes Back" : True,
    "Star Wars: Episode VI Return of the Jedi" : True,
    "No" : False
}
```

```python
star_wars.head(2)
```

```python
star_wars["Which of the following Star Wars films have you seen? Please select all that apply."]=star_wars["Which of the following Star Wars films have you seen? Please select all that apply."].map(movies)
star_wars["Unnamed: 4"]=star_wars["Unnamed: 4"].map(movies)
star_wars["Unnamed: 5"]=star_wars["Unnamed: 5"].map(movies)
star_wars["Unnamed: 6"]=star_wars["Unnamed: 6"].map(movies)
star_wars["Unnamed: 7"]=star_wars["Unnamed: 7"].map(movies)
star_wars["Unnamed: 8"]=star_wars["Unnamed: 8"].map(movies)
```

```python
column_names = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4" : "seen_2",
    "Unnamed: 5" : "seen_3",
    "Unnamed: 6" : "seen_4",
    "Unnamed: 7" : "seen_5",
    "Unnamed: 8" : "seen_6",
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10" : "ranking_2",
    "Unnamed: 11" : "ranking_3",
    "Unnamed: 12" : "ranking_4",
    "Unnamed: 13" : "ranking_5",
    "Unnamed: 14" : "ranking_6",
}
```

```python
by_age = star_wars.pivot_table(index = 'Age')
```

```python
fig = plt.figure(figsize = (10,12))
ax1 = fig.add_subplot(2,1,1)
count = 0
color_age = ['red','blue','green','yellow']
c_i = 0
ar = np.arange(1.3,7.0,1.0)
a = []
for age in by_age.index.values:
    ratings = by_age[by_age.columns[2:8]].loc[age].values
    a.append(ax1.bar(ar,ratings,0.10,color = color_age[c_i]))
    c_i+=1
    ar += 0.1
ax1.set_xticks(np.arange(1.5,7,1))
ax1.set_xticklabels(by_age.columns[2:8],rotation=30)
ax1.legend(a,['18-29','30-44','45-60','> 60'])
ax1.set_title('movie ranking by age')
##############################
ax2 = fig.add_subplot(2,1,2)
count = 0
color_age = ['red','blue','green','yellow','violet','orange']
c_i = 0
ar = np.arange(1.2,5.0,1.0)
a2 = []
for rank in by_age.columns[2:8].values:
    ratings = by_age[by_age.columns[2:8]].transpose().loc[rank].values
    a2.append(ax2.bar(ar,ratings,0.10,color = color_age[c_i]))
    c_i+=1
    ar += 0.1
ax2.set_xticks(np.arange(1.5,5,1.0))
ax2.set_xticklabels(by_age.transpose().columns,rotation=0)
ax2.legend(a2,list(by_age.columns[2:8]))
ax2.set_title('rankings by age')
plt.show()
```

```python
columns = list(star_wars.columns[3:9])
columns.append('Age')
seen_by_age = by_age = star_wars[columns].pivot_table(index='Age')
```

```python
age_labels = by_age['Age'].unique()
```

```python
age_labels = age_labels.take([0,2,3,4])
```

```python
age_labels
```

```python
seen_by_age = [(by_age[by_age.columns[:6]][by_age['Age']==x]).sum() for  x in (age_labels)]
```

```python
by_age[by_age.columns[:6]][by_age['Age']=='45-60'].sum()
```

```python
seen_by_age
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ticks = np.arange(1.3,7,1)
colors = ['red','blue','yellow','green']
count = np.zeros(6);
bars = list()
for j,by_age in enumerate(seen_by_age):  
    bars.append(ax.bar(np.arange(1.1,7,1),by_age,color = colors[j],bottom= count))
    count += by_age;
ax.set_xticks(np.arange(1.5,7,1))
ax.set_xlim(1,7)
ax.set_xticklabels(by_age.index,rotation=30)
ax.set_ylim(0,900)
ax.legend(bars,list(age_labels))
ax.set_title('Views per movies per ages')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ticks = np.arange(1.3,7,1)
colors = ['red','blue','yellow','green']
count = np.arange(1.1,7.0,1.0)
bars = list()
for j,by_age in enumerate(seen_by_age):  
    bars.append(ax.bar(count,by_age,0.2,color = colors[j]))
    count += 0.2;
ax.set_ylim(0,250)
ax.set_xticks(np.arange(1.5,7,1))
ax.set_xlim(1,7)
ax.set_xticklabels(by_age.index,rotation=30)
ax.legend(bars,list(age_labels))
ax.set_title('movie views per age')
plt.show()
```

We can see that great majority of the survey people are older than 30 years. People older than 30 years  grown up seeing Star Wars Episodes IV, V and VI (in that time where the original first three),  

The most view movie is movie Episode 5: The Empire Strikes Back

```python
males = star_wars[star_wars['Gender']=='Male']
females = star_wars[star_wars['Gender']=='Female']
```

```python
sw_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
sw_nofan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]
```

```python
st_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'Yes']
st_nofan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'No']
```

```python
males_seens = males[star_wars.columns[3:9]]
mMovie_count = males_seens.sum()
females_seens = females[star_wars.columns[3:9]]
fMovies_count = females_seens.sum()
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(1,1,1)
ar = np.arange(1.1,7,1.0)
a1= ax.bar(ar,fMovies_count,0.4,color="red")
ar+=0.4
a2= ax.bar(ar,mMovie_count,0.4,color="blue")
ax.set_ylim(0,450)
ax.legend([a1,a2],['Female','Male'])
ax.set_xticks(np.arange(1.5,7.0,1.0))
ax.set_xticklabels(star_wars.columns[3:9],rotation=30)
ax.set_title("movie view count by gender")
plt.show()
```

```python
swFan_seens = sw_fan[star_wars.columns[3:9]]
swFan_count = swFan_seens.sum()
swNFan_seens = sw_nofan[star_wars.columns[3:9]]
swNFan_count = swNFan_seens.sum()
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(1,1,1)
ar = np.arange(1.1,7,1.0)
a1= ax.bar(ar,swFan_count,0.4,color="red")
ar+=0.4
a2= ax.bar(ar,swNFan_count,0.4,color="blue")
ax.set_ylim(0,650)
ax.legend([a1,a2],['Fan','No Fan'])
ax.set_xticks(np.arange(1.5,7.0,1.0))
ax.set_xticklabels(star_wars.columns[3:9],rotation=30)
ax.set_title("movie view count by SW fan")
plt.show()
```

```python
stFan_seens = st_fan[star_wars.columns[3:9]]
stFan_count = stFan_seens.sum()
stNFan_seens = st_nofan[star_wars.columns[3:9]]
stNFan_count = stNFan_seens.sum()
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(1,1,1)
ar = np.arange(1.1,7,1.0)
a1= ax.bar(ar,stFan_count,0.4,color="red")
ar+=0.4
a2= ax.bar(ar,stNFan_count,0.4,color="blue")
ax.set_ylim(0,650)
ax.legend([a1,a2],['Fan','No Fan'])
ax.set_xticks(np.arange(1.5,7.0,1.0))
ax.set_xticklabels(star_wars.columns[3:9],rotation=30)
ax.set_title("movie view count by St fan")
plt.show()
```

when we see the last three plots and the plot about movie's ranking, we can see there is a correlation. The best ranking movie is Episode V, and in these plots the most seen movie is Episode V. There is a correlation between the movie's ranking and the movie's seen count; movie's seen count growns respect the ranking of the movie for all this groups
